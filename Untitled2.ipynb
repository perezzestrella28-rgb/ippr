```python
#calculadora IMC
peso = float(input("ingresar tu peso"))
altura = float(input("ingresar tu altura"))

imc = peso / (altura ** 2)
print(f'tu IMC es: {imc:2f}')

if imc < 18.5:
   print("bajo peso")
elif imc < 25:
  print("peso normal")

elif imc < 30:
  print("sobrepeso")

else:
  print("eres un kilo mortal")
```
