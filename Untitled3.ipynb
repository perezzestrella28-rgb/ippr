```python
#operacion con listas
#ESTRELLA ISABEL

frutas = ['manzana', 'pera', 'uva', 'mango']

print(frutas[0]) #mostrar la primera fruta
print(frutas[-1]) #mostrar la ultima fruta

frutas[1] = 'naranja' #notificacion de posicion 1
frutas.append('kiwi') #agregar al final
frutas.insert(1, 'fresa') #insertar en posicion 1
frutas.remove('uva') #eliminar la 1era
frutas.pop(0)
```

```python
#lista con numeros
#ESTRELLA ISABEL

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]
print('original:', numeros)

# ordenar de menor a mayor
numeros.sort()
print('ordenada:', numeros)

# ordenar de mayor a menor
numeros.reverse()
print('invertida:', numeros)
```

```python
#promedio
#ESTRELLA ISABEL

numeros = [45, 12, 17, 78, 3, 56, 89, 23, 67, 1, 34]
total = sum(numeros)
promedio = sum(numeros) / len(numeros)

print(f'suma: {total}')
print(f'promedio: {promedio: .2f}')
```
